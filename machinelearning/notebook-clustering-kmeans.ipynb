## A toy example of Kmeans clustering
1. Generate two Gaussian distributed clusters
2. Unsupervisedly clustering with Kmeans method

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### Randomly generate two clusters with 2D Gaussian distribution

```python
mu1_x = 5.0
sigma1_x = 2.0
mu1_y = 10.0
sigma1_y = 1.0
numpoints1 = 200
points_c1 = np.random.multivariate_normal([mu1_x,mu1_y], np.array([[sigma1_x, 0], [0, sigma1_y]]),size=numpoints1)
```

```python
mu2_x = 6.0
sigma2_x = 1.0
mu2_y = 4.0
sigma2_y = 3.0
numpoints2 = 200
points_c2 = np.random.multivariate_normal([mu2_x,mu2_y], np.array([[sigma2_x, 0], [0, sigma2_y]]),size=numpoints2)
```

### Clustering

```python
from sklearn.cluster import KMeans
```

```python
#x_train = np.hstack([points_c1[:,0], points_c2[:,1]])
#y_train = np.hstack([points_c2[:,0], points_c2[:,1]])
data_train = np.vstack([points_c1, points_c2])
```

```python
random_state = 170
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(data_train)
```

### Display the result

```python
plt.rcParams["figure.figsize"] = [12.0, 5.0]
ax1 = plt.subplot(121)
ax1.scatter(points_c1[:,0], points_c1[:,1], marker='^')
ax1.scatter(points_c2[:,0], points_c2[:,1], marker='o')

ax2 = plt.subplot(122)
#axis_c1 = np.where(y_pred == 0)
#axis_c2 = np.where(y_pred == 1)
#ax2.scatter(data_train[axis_c1,0],data_train[axis_c1,1], marker='^')
#ax2.scatter(data_train[axis_c2,0],data_train[axis_c2,1], marker='o')
ax2.scatter(data_train[:,0], data_train[:,1], c=y_pred)
```
